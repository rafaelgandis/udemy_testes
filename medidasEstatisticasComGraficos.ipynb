```python
import statistics
from matplotlib import pyplot as plt
```

```python
numeros = [10, 20, 20, 20, 30, 40, 50, 60]
```

```python

def calcular_estatisticas(numeros):

    estatisticas = {}

    estatisticas['media'] = statistics.mean(numeros)

    estatisticas['mediana'] = statistics.median(numeros)

    try:
        estatisticas['moda'] = statistics.mode(numeros)
    except statistics.StatisticsError:
        estatisticas['moda'] = "No unique mode"

    estatisticas['variancia'] = statistics.variance(numeros)

    estatisticas['desvio_padrao'] = statistics.stdev(numeros)

    return estatisticas


estatisticas = calcular_estatisticas(numeros)
print(estatisticas)
```

```python
def plot_statistics (estatisticas):
    categorias = list(estatisticas.keys())
    valores = list(estatisticas.values())
    plt.figure(figsize=(10, 6))
    plt.figure(figsize=(10, 6))
    plt.bar(categorias, valores, color=['blue', 'green', 'red', 'purple', 'orange'])
    plt.xlabel('Estatísticas')
    plt.ylabel('Valores')
    plt.title('Medidas Estatísticas')
    plt.show()
```

```python
estatisticas = calcular_estatisticas(numeros)
plot_statistics (estatisticas)
```
